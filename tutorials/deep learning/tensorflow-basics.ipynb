```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

a = np.array([1,2])
b = np.array([3,4])
c= a+b
print(c)
```

### Add , Minus

```python
a = tf.constant([1,2])
b = tf.constant([3,4])
c = a+b
d = a-b
print(c,d)
```

### Multiply

```python
a = tf.constant([[1,2], [3,5]])
b = tf.constant([[0,4], [1,3]])
c = a*b # tf.multiply(a,b)  # Element multiplication
d = tf.matmul(a, b)         # Matrix multiplication
print(c, d)
```

### Division

```python
a = tf.constant([[1,2], [3,5]])
b = tf.constant([[0,4], [1,3]])
c = a/b
d= tf.divide(a,b)
print(c,d)
```

### Reshape

```python
a = tf.constant([1,2,3,4,5,6])
b = tf.reshape(a,[2,3])
print(a,b)
```

### Min, Max, Sum, Mean, STD

```python
a = tf.constant([[1,3], [2,5]], dtype=float)
min = tf.reduce_min(a)
max = tf.reduce_max(a)
sum = tf.reduce_sum(a)
mean = tf.reduce_mean(a)
std = tf.math.reduce_std(a)
print("min : ",min)
print("max : ",max)
print("sum : ",sum)
print("mean : ",mean)
print("std : ",std)
```

## Simple Neural Network

```python
import keras

#  function we want to find : y = 3x + 4
x_train = np.array([2,4,7,8,12,22,10])
x_test = np.array([1,9,15])
y_train = np.array([10,16,25,28,40,70,34])
y_test = np.array([7, 31,49])

model_simple = keras.Sequential([keras.layers.Dense(units=1, activation=None, input_shape=[1])])
model_simple.compile(optimizer="sgd", loss="mean_squared_error", metrics=['accuracy'])
model_simple.summary()
```

```python
model_simple.fit(x_train, y_train, epochs=200, batch_size=1)
```

```python
predict = model_simple.predict(x_test)
print("targets  : ", y_test)
print("predicts : ", predict)
```

## 1x1024 Neuron Layer vs 4x256 Neuron Layer 

#### create dataset

```python
from sklearn.model_selection import train_test_split
x = np.arange(-80,81, step=0.5)
y= x**2 + 5*x + 3
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
print(len(x),len(x_train), len(x_test))
```

#### 1x1024 model

model_1 = keras.Sequential([])
model_1.add(keras.layers.Dense(units=512, activation="relu", input_shape=[1]))
model_1.add(keras.layers.Dense(units=512, activation="relu"))
model_1.add(keras.layers.Dense(units=1, activation="relu"))
model_1.compile(optimizer="adam", loss=keras.losses.MeanSquaredError)
model_1.summary()

```python
model_1.fit(x_train,y_train,batch_size=16,epochs=100,validation_data=(x_test,y_test))
```

#### 4x256 model

```python
model_2 = keras.Sequential([])
model_2.add(keras.layers.Dense(units=256, activation="relu", input_shape=[1]))
model_2.add(keras.layers.Dense(units=256, activation="relu"))
model_2.add(keras.layers.Dense(units=256, activation="relu"))
model_2.add(keras.layers.Dense(units=256, activation="relu"))
model_2.add(keras.layers.Dense(units=1, activation="relu"))
model_2.compile(optimizer="adam", loss="mean_squared_error")
model_2.summary()
```

```python
model_2.fit(x_train,y_train,batch_size=16,epochs=100,validation_data=(x_test,y_test))
```

#### Manual model. a Human examine data
we seperate data base on function going up or down

```python
# find function minimum value using derivative x**2 + 5x +3  => dy/dx = 2x + 5 = 0 => x= -2.5
mask = x_train< -2.5
x_1 = x_train[mask]
x_2 = x_train[~mask]
y_1 = y_train[mask]
y_2 = y_train[~mask]

model_m_1 = keras.Sequential([])
model_m_1.add(keras.layers.Dense(units=256, activation="relu", input_shape=[1]))
model_m_1.add(keras.layers.Dense(units=1, activation="relu"))
model_m_1.compile(optimizer="adam", loss="mean_squared_error")
model_m_1.fit(x_1, y_1,batch_size=16, epochs=200)

model_m_2 = keras.Sequential([])
model_m_2.add(keras.layers.Dense(units=256, activation="relu", input_shape=[1]))
model_m_2.add(keras.layers.Dense(units=1, activation="relu"))
model_m_2.compile(optimizer="adam", loss="mean_squared_error")
model_m_2.fit(x_2, y_2,batch_size=16, epochs=200)
```

```python
def manual_model_predict(input:list[int]) -> list[int]:
    mask = input < -2.5
    input_1 = input[mask]
    input_2 = input[~mask]
    pred_1 = model_m_1.predict(input_1)
    pred_2 = model_m_2.predict(input_2)
    predicts = np.concatenate((pred_1,pred_2))
    return predicts
```

```python
model_1_predicts = model_1.predict(x_test)
model_2_predicts = model_2.predict(x_test)
model_m_2_p = model_m_2.predict(x_test)
model_manual_predicts = manual_model_predict(x_test)

# print("targets          : ", y_test)
# print("model_1_predicts : ", model_1_predicts)
# print("model_2_predicts : ", model_2_predicts)
# print("model_manual_predicts : ", model_manual_predicts)

plt.scatter(x_test, y_test, c= "black", linewidths=10)
plt.scatter(x_test, model_2_predicts, c= "blue", linewidths=7)
plt.scatter(x_test, model_m_2_p, c= "yellow", linewidths=4)
plt.scatter(x_test, model_1_predicts, c= "red", linewidths=3)
plt.show()
```

#### 1x1024 NN to predict a quadratic func

```python
x = np.arange(-80, 81, step=0.5)
y = x**2 + 5*x + 3

x_norm = x / np.max(x) 
y_norm = y / np.max(y) 

x_train, x_test, y_train, y_test = train_test_split(x_norm,y_norm,test_size=0.2)

model_simple = keras.Sequential([])
model_simple.add(keras.layers.Dense(units=1024, activation="sigmoid", input_shape=[1]))
model_simple.add(keras.layers.Dense(units=1, activation="sigmoid"))
model_simple.compile(optimizer="adam", loss="mse")
model_simple.fit(x_train, y_train, batch_size=4, epochs=150)

y_train_predict = model_simple.predict(x_train)
y_test_predict = model_simple.predict(x_test)

x_train_unnorm = x_train * np.max(x)
y_train_unnorm = y_train * np.max(y)
y_train_predict_unnorm = y_train_predict * np.max(y)

x_test_unnorm = x_test * np.max(x)
y_test_unnorm = y_test * np.max(y)
y_test_predict_unnorm = y_test_predict * np.max(y)


plt.scatter(x_train_unnorm, y_train_unnorm, c="blue", linewidths=7)
plt.scatter(x_train_unnorm, y_train_predict_unnorm, c="red", linewidths=3)
plt.show()
```
