### Data Set

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

data_list , data_target = make_blobs(n_samples=300,n_features=2, centers=3)
x = data_list[:,0]
y = data_list[:,1]
plt.scatter(x,y)
```

```python
from sklearn.cluster import KMeans

model = KMeans(3)
model.fit(data_list)
```

```python
labels = model.labels_
centers = model.cluster_centers_

plt.scatter(x,y,c=labels)
plt.scatter(centers[:,0],centers[:,1], marker="X")

# cluster_count = 3
# masks = []
# for cluster_idx in range(cluster_count):
#     mask = [i == cluster_idx for i in labels]
#     masks.append(mask)

# for cluster_idx in range(cluster_count):
#     color = "red"
#     if cluster_idx ==1: color = "blue"
#     if cluster_idx ==2: color = "black"
#     plt.scatter(x[masks[cluster_idx]], y[masks[cluster_idx]],c=color)
#     plt.scatter(centers[cluster_idx][0], centers[cluster_idx][1], c="green", marker="s")
```
